```python
lst = []
for i in df['Age']:
  if i >=18 and i<=25:
    lst.append('Young Adults')
  elif i>25 and i<=35:
    lst.append('Early Career')
  elif i>35 and i<=45:
    lst.append('Mid Career')
  elif i>45 and i<=55:
    lst.append('Late Career')
  elif i>55 and i<=64:
    lst.append('Pre-retirement')
  else:
    lst.append(False)
```

```python
df['Age Group' ] = lst
```

```python
corrections = {'Barzil':'Brazil'}
df['Location'].replace(corrections , inplace=True)
```

## UNIVARIATE ANALYSIS

```python
sns.countplot(data = df, x = 'Age Group',hue='Gender',palette= 'Blues')
plt.title('Age Group Distribution')
plt.show()
#It seems that the people of "Mid Career" and "Late Career" spends more time on social media.
```

```python
sns.countplot(data = df , x = 'Gender')
plt.title('Distinguish According to Gender')
plt.show()
#According to Gender , "Male" shows most interest in social media
```

```python
plt.figure(figsize = (8,5))
sns.countplot(data = df , x = 'Location',hue='Demographics' ,palette='coolwarm')
plt.xticks(rotation = 70)
plt.show()
#According to this data , the people of rural India spends more time on social media while the people of Germany spends less.
```

```python
# Operations Team: Analyzing Total Time Spent and Video Category Engagement
plt.figure(figsize=(6, 6))
sns.barplot(x='Video Category', y='Total Time Spent', data=df,color='yellow')
plt.title('Total Time Spent per Video Category')
plt.xticks(rotation=45)
plt.show()
```
